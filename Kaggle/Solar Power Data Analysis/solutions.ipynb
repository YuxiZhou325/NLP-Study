<h2>Task 1.1</h2>
identify the percentage of missing data for each inverter in the column "TOTAL_YIELD"

```python
# solution of task 1.1
import pandas as pd

# read csv data file with pandas
df = pd.read_csv('G:\JobApply\BayWaRe\Plant_1_Generation_Data.csv')

# view the data frame
df.head()
```

```python
# view TOTAL_YIELD column
df.TOTAL_YIELD
```

```python
# check if there is missing values in TOTAL_YIELD column
df['TOTAL_YIELD'].isnull().any()
```

```python
# check if there is 0 value in TOTAL_YIELD column
# if it returns false, then means there are at least one Zero value in the field
df['TOTAL_YIELD'].all().any()
```

```python
# get the count of 0 in TOTAL_YIELD column
count = (df['TOTAL_YIELD'] == 0).sum()
print("The count of Zeros in TOTAL_YIELD field is: ", count)
```

```python
# group the dataframe according to the inverter id
grouped_df = df.groupby('SOURCE_KEY', as_index='False')

# df['TOTAL_YIELD'].isnull() / df.groupby('SOURCE_KEY')['TOTAL_YIELD'].transform('sum')

# calculate the total rows for each inverter
grouped_df['TOTAL_YIELD'].count()
```

```python
# view the amount of missing values by each inverter through the whole data frame
df.isnull().groupby(df.SOURCE_KEY, sort=False).sum()
```

```python
# view how many missing values in TOTAL_YIELD by each inverter
df['TOTAL_YIELD'].isnull().groupby(df.SOURCE_KEY, sort=False).sum()
```

<h4>Sub Conclusion</h4>
The results indicated that there is no missing value and Zeros in the TOTAL_YIELD field

```python
# Calculate the % of missing data for each inverter
df['TOTAL_YIELD'].isnull().groupby(df.SOURCE_KEY, sort=False).sum() / grouped_df['TOTAL_YIELD'].count()
```

<h2>Task 1.3</h2>
calculate the production of each inverter on a 15min basis and visualise the result

```python
# view the production for each 15 min time slot
grouped_df['TOTAL_YIELD'].diff()
```

```python
import matplotlib.pyplot as plt

# calculate the every 15min production for each inverter by diff()
# within the grouped total yield by inverter ids
# plot the result in one png figure to compare and analysis

# set figure size
plt.figure(figsize=(60,20))
for key, item in grouped_df['TOTAL_YIELD']:
    # print(grouped_df['TOTAL_YIELD'].get_group(key).diff(), "\n\n")
    plt.title("Production of Each Inverter on a 15min Basis During 34 Days")
    # draw the 15min basis production for each inverter
    plt.plot(grouped_df['TOTAL_YIELD'].get_group(key).diff())
    # plt.legend(loc='best')
# plt.show()
plt.savefig('15min_production_for_each_inverter.png')
```
